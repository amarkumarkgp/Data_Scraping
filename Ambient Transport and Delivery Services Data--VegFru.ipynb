```python
import lxml
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd 
import numpy as np
```

```python
f_data =[] 
f_data = pd.DataFrame(f_data)
for k in range(0,30,1):
    k = str(k)
    if k =='0':
        url = r'https://vegfru.com/service-search/distribution_and_logistics_services/ambient_transport_and_delivery_services?sub_category=ambient_transport_and_delivery_services'
    elif k =='14':
        continue
    else:
        url = r'https://vegfru.com/service-search/'+k+'/distribution_and_logistics_services/ambient_transport_and_delivery_services?sub_category=ambient_transport_and_delivery_services'
    
    source = requests.get(url).text
    soup = bs(source,'lxml')
    #htmllist_2 = soup.select(".col-md-30.listing-cat")

    htmllist_a = soup.select(".pure-u-1-1.pure-u-md-18-24")
    htmllist_b = soup.select(".col-md-70.expertise-detail")
    corpus_a =[]
    for i in range(0,len(htmllist_a)):
        temp = re.sub('[^a-zA-Z0-9]', ' ', str(htmllist_a[i].text))
        temp = temp.split()
        temp = ' '.join(temp)
        corpus_a.append(temp)
    corpus_b =[]
    for i in range(0,len(htmllist_b)):
        temp = re.sub('[^a-zA-Z0-9]', ' ', str(htmllist_b[i].text))
        temp = temp.split()
        temp = ' '.join(temp)
        corpus_b.append(temp)
    
    
    from numpy import *
    datas = reshape(corpus_b,(10,5))
    datas = pd.DataFrame(datas)
    
    edata  = []
    edata = pd.DataFrame(edata)
    edata['Type'] = corpus_a[2:]
    edata["Specialty Area"] = datas.iloc[:,0]
    edata["Product Specialty"] =datas.iloc[:,1]
    edata["No. Of Vehicles"] =datas.iloc[:,2]
    edata["Type Of Vehicles"]=datas.iloc[:,3]
    edata["Permit"]=datas.iloc[:,4]
    
    f_data = f_data.append(other = edata, ignore_index=True)
    
    
    
```
